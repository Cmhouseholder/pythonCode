    DNA (1:10) dilution = 2 uL
    Water = 10.4 uL
    F primer (10 uL) = 1 uL
    R primer (10 uL) = 1 uL
    M13-FAM primer (10 uM) = 1 uL

    5x buffer = 4 uL

```python
# calculate volumes
#######
genes=[];

mygene1 = 'tbx2a'
gene_samplenum = 14
control_num = 3

genes.append(gene(mygene1,(array2list(np.arange(0,gene_samplenum+control_num)+1)))); # the +1 after gene len1 is for control.
# genes.append(gene('isl1a',(array2list(np.arange(0,8*10)+1))));


# genes.append(gene('eml1ii',(array2list(np.arange(0,12,1)))));

PCR = pd.DataFrame.copy(SolutionMix)
for g in genes:
    PCR[g.string] = round(pd.Series(PCR['1x'] * (g.n*1*1.1)),2)
PCR.columns.name = '-'
PCR.round(decimals=2)
```

```python
# generate sample IDs
names = [];
ids = [];
tms = [];
seqPrimer =[];
expectedSize =[];
for g in genes:
    names.append(g.name)
    ids.append(g.ids)
    tms.append(g.Tm)
    seqPrimer.append(g.seqPrimer)
    expectedSize.append(g.expectedSize)

IDs = pd.DataFrame({'id#': ids, 'Tm (°C)' : tms, 'seqPrimer' : seqPrimer, 'expectedSize (bp)' : expectedSize}, index=names)
IDs.columns.name = 'gene'
#IDs[['id#','Tm (°C)']]
IDs
```
